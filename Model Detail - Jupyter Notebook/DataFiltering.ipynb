```python
import numpy as np
import pandas as pd
```

```python
mumbai_data = pd.read_csv("Mumbai1_cleaned.csv")
mumbai_data.insert(loc=0, column='Price per sqft', value=None)
mumbai_data['Price per sqft'] = mumbai_data['Price']/mumbai_data['Area']
# mumbai_data.drop(columns=['Price',] , inplace = True)
```

```python
# pd.DataFrame(mumbai_data['Location'].value_counts()).to_csv("Location Frequency.csv")
locationF = mumbai_data['Location'].value_counts()
```

```python
def remove_outliers_grouped(df, group_col, value_col):
    df_no_outliers = pd.DataFrame()
    for name, group in df.groupby(group_col):
        Q1 = group[value_col].quantile(0.25)
        Q3 = group[value_col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        filtered_group = group[(group[value_col] >= lower_bound) & (group[value_col] <= upper_bound)]
        df_no_outliers = pd.concat([df_no_outliers, filtered_group])
    return df_no_outliers

mumbai_data = remove_outliers_grouped(mumbai_data, 'Location', 'Price per sqft')
mumbai_data = remove_outliers_grouped(mumbai_data, 'Location', 'Area')
```

```python
mumbai_data = mumbai_data[mumbai_data['Location'].isin(locationF[locationF>=50].index)]
avg_land_price_by_location = mumbai_data.groupby('Location')['Price'].mean()/mumbai_data.groupby('Location')['Area'].mean()
# avg_land_price_by_location = pd.DataFrame(avg_land_price_by_location)
# avg_land_price_by_location.columns = ['Average Price of 1sqFt']
```

```python
mumbai_data['Average_price_per_sq_ft'] = mumbai_data['Location'].map(avg_land_price_by_location)
```

```python
#Performing One Hot Encoding on the Location data
mumbai_data_encoded = pd.get_dummies(mumbai_data , columns=['Location'] , drop_first=False )
```

```python
'''
The encoding created the columns of diffrent location and named them like "Location_....."
we want to remove this "Location_" so that it become easy to read
'''
for columns in mumbai_data_encoded.columns:
    if(columns.startswith('Location_')):
        mumbai_data_encoded.rename(columns={columns: columns[9:]} , inplace=True)
```

```python
'''
Our Dataset have Encoded boolean values in the form of True/False.
for this data to be fitted into a model we need to convert this into 1/0.
'''
mumbai_data_encoded
```

```python
mumbai_data_encoded.info()
```

```python
'''
we want to convert Boolean Data (True , False) to their integer form (1,0).
So that it can be fitted into our model
'''
from pandas.api.types import is_bool_dtype
for columns in mumbai_data_encoded.columns:
    if(is_bool_dtype(mumbai_data_encoded[columns])):
        mumbai_data_encoded[columns]= mumbai_data_encoded[columns].astype(int)
```
